```python
#I created a copy for testing and ease of use.
testingData = busData.copy()

#Converts the data's time column to datetime format for easier comparisons.
testingData['Aeg'] = pd.to_datetime(testingData['Aeg']) 

```

```python
#Extracting certain days for graphing.
metallicaDayData = testingData[(testingData['Aeg'].dt.day == 18) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour >= 12) & (testingData['Aeg'].dt.hour > 0)].append(testingData[(testingData['Aeg'].dt.day == 19) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour < 12) & (testingData['Aeg'].dt.hour >= 0)])
punkData = testingData[(testingData['Aeg'].dt.day == 13) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour >= 12)].append(testingData[(testingData['Aeg'].dt.day == 14) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour < 12)])
randomSaturday = testingData[(testingData['Aeg'].dt.day == 27) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour >= 12)].append(testingData[(testingData['Aeg'].dt.day == 28) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour < 12)])
randomThursday = testingData[(testingData['Aeg'].dt.day == 4) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour >= 12) & (testingData['Aeg'].dt.hour > 0)].append(testingData[(testingData['Aeg'].dt.day == 5) & (testingData['Aeg'].dt.month == 7) & (testingData['Aeg'].dt.hour < 12) & (testingData['Aeg'].dt.hour >= 0)])

#Not that important to sort the bus lines and directions, but done just in case.
metallicaDayData.sort_values(["Liin","Suund","Aeg"])
punkData.sort_values(["Liin","Suund","Aeg"])
randomThursday.sort_values(["Liin","Suund","Aeg"])
metallicaDayData
```

```python
#For removing the lines that don't have much usage or stop driving after 15/16 o'clock, as it will not give much information and just make the final graph more confusing.
def removeUselessLines(data):
    BusLinesToRemove = ["9","9A",21,22,25]
    for line in BusLinesToRemove:
        data = data[data.Liin != line]
    return data

#For sorting the bus lines, function made because busline lists usually contain some string values. If reverse is true, the list will be sorted in descending order.
def sortBusLines(data,reverse):
    n = len(data)
 
    for i in range(n):
 
       
        for j in range(0, n-i-1):
 
            if int(data[j]) > int(data[j+1]) :
                data[j], data[j+1] = data[j+1], data[j]
            
    if (reverse):
        data.reverse()
        
                
def divideDataTicket(dayData):
    
    ticket_names = metallicaDayData["Toote nimetus"].value_counts().index.tolist()
    ticket_names.reverse()
    ticket_order = pd.api.types.CategoricalDtype(categories=ticket_names, ordered=True)
    print(len(ticket_names))
    dayData = removeUselessLines(dayData)
    
    busLines = dayData["Liin"].value_counts().index.tolist()
    
    sortBusLines(busLines,True)
    
    ticketData = []
    
    
    for line in busLines:
        ticketData.append(dayData[(dayData)["Liin"] == line]['Toote nimetus'].astype(ticket_order).value_counts().sort_index())
        
    
    
    return ticketData,busLines  
            
#For dividing a day's data into hour long sections. Returns a 2D array, where each row is for a busline and each column element for a subdataframe that represents that hours data, and the correct buslines.            
def divideData(dayData):
    
    dayData = removeUselessLines(dayData)
    
    busLines = dayData["Liin"].value_counts().index.tolist()
    
    sortBusLines(busLines,True)
    
    #Creates empty nested list with size of busLines.
    hourlyValidationData = [[] for _ in range(len(busLines))]
    
    index = 0

    for line in busLines:
        
        #This loop goes thru 12 to 23
        for i in range(12,24):
            
            #Finds the data where the line is same as the first loops line value and the hour is between i and i+1.
            temp1 = dayData[(dayData["Liin"] == line) & (dayData["Aeg"].dt.hour >= i) & (dayData["Aeg"].dt.hour < (i+1))]
            
            #If in the certain hour there is no data, then just put in none and keep going
            if(temp1.shape[0] == 0):
                hourlyValidationData[index].append(None)
                continue
            
            #Otherwise, append that subdataframe to the list.
            hourlyValidationData[index].append(temp1)

        #This loop goes thru 0 to 11. Basically the same thing as the earlier loop, but does it for the morning.
        for i in range(0,2):
            
            #Ignore hours 2-3,3-4 and 4-5, as almost no busses go at that time and it would make the end graph more confusing.
            if (i in [2,3,4]):
                continue
            temp2 = dayData[(dayData["Liin"] == line) & (dayData["Aeg"].dt.hour >= i) & (dayData["Aeg"].dt.hour < (i+1))]
            if(temp2.shape[0] == 0):
                hourlyValidationData[index].append(None)
                continue
            hourlyValidationData[index].append(temp2)

        index += 1
        
    return hourlyValidationData,busLines

#Returns a 2D array of how many validations were made in each hour on each bus line. Needed for the "z" axis data.
def getValidationCount(hourlyData, busLines): 
    validationValues = [[] for _ in range(len(busLines))]

    for i in range(len(validationValues)):
        for j in range(len(hourlyData[i])):
            try:
                #try to get a count of how many passangers validated in that hour.
                validationValues[i].append(hourlyData[i][j].Reisijaid.count())
            except:
                #if in that hour there was no data(None) then just append another None for the final graph.
                validationValues[i].append(None)
    return validationValues

#Finds the correct coordinates for the bar chart.
def getCoordinates(validationValues, colors):
    
    validationValues
    
    for i in range(11):
        print(validationValues[i])
        print()
    
    x_pos = []
    y_pos = []
    colorsForGraph = []
    #These are basically the coordinates for the graph, these are to be appended in the correct order to the previous lists.
    x_pos_counter = 0
    y_pos_counter = 0


    for i in range(len(validationValues)):

        for j in range (len(validationValues[i])):
            #If there is no data on that certain hour, then skip ahead to the next y position.
            if (validationValues[i][j] == None):
                y_pos_counter += 1
                continue
            #Otherwise, add the coordinate to the List.
            x_pos.append(x_pos_counter)
            y_pos.append(y_pos_counter)
            colorsForGraph.append(colors[x_pos_counter])
            y_pos_counter += 1
            
        #Reset the y counter as the next line starts again from the first position(from the front of the screen so to say) and go to the next x position (next busline)
        y_pos_counter = 0
        x_pos_counter += 1
        
    return(x_pos,y_pos,colorsForGraph)  

def getCoordinatesTickets(ticketData,ticketNames, colors):
 
    x_pos = []
    y_pos = []
    
    colorsForGraph = []

    #These are basically the coordinates for the graph, these are to be appended in the correct order to the previous lists.
    x_pos_counter = 0
    y_pos_counter = 0


    for i in range(len(ticketData)):

        for j in range (len(ticketNames)):
            if(ticketData[i][j] == 0):
                y_pos_counter += 1
                continue
            x_pos.append(x_pos_counter)
            y_pos.append(y_pos_counter)
            colorsForGraph.append(colors[x_pos_counter])
            y_pos_counter += 1
            
        #Reset the y counter as the next line starts again from the first position(from the front of the screen so to say) and go to the next x position (next busline)
        y_pos_counter = 0
        x_pos_counter += 1
        
    return(x_pos,y_pos,colorsForGraph) 

#For getting the correct positions for "z" axis values.
def getZValues(validationValues):
    z_Values = []
    #loop goes thru all bus lines and non None values in the hourly validation Values.
    for busLine in validationValues:
        busLineWithoutNones = []
        
        for value in busLine:
            if (value != None):
                busLineWithoutNones.append(value)
        
        for value in busLineWithoutNones:
            z_Values.append(value)
            
    return z_Values


def getTicketZValues(ticketData):
    ticket_values = []
    
    for busLine in ticketData:
        for value in busLine.tolist():
            if(value >0):
                ticket_values.append(value)
    return ticket_values
```

```python
row_names = ["12-13", "13-14", "14-15", "15-16", "16-17", "17-18", "18-19", "19-20", "20-21", "21-22", "22-23", "23-24", "00-01", "01-02" ]
ticket_names = metallicaDayData["Toote nimetus"].value_counts().index.tolist()
listOfColors = ['#f032e6','#911eb4','#4363d8','#42d4f4','#ffe119','#f58231','#e6194b','#469990', '#808000', '#9a6324', '#800000']


metallicaHourly, metallicaBusLines = divideData(metallicaDayData)

metallicaValidationValues = getValidationCount(metallicaHourly,metallicaBusLines)

met_z_values = getZValues(metallicaValidationValues)

met_x_pos,met_y_pos,met_colors = getCoordinates(metallicaValidationValues,listOfColors)

met_z = np.zeros(len(met_x_pos))  #z coordinates of each bar


met_x = np.arange(len(metallicaBusLines))

met_y = np.arange(len(row_names))

```

```python
normalHourly, normalBusLines = divideData(randomThursday)

normalValidationValues = getValidationCount(normalHourly,normalBusLines)

normal_z_values = getZValues(normalValidationValues)

normal_x_pos, normal_y_pos,n_colors = getCoordinates(normalValidationValues,listOfColors)

normal_z = np.zeros(len(normal_x_pos))

normal_x = np.arange(len(normalBusLines))

normal_y =  np.arange(len(row_names))
```

```python

met_ticketData,buslines = divideDataTicket(metallicaDayData)

ticket_x,ticket_y,ticket_colors = getCoordinatesTickets(met_ticketData,ticket_names,listOfColors)

ticket_z = getTicketZValues(met_ticketData)

```

```python
normal_ticketData, buslines = divideDataTicket(randomThursday)
n_ticket_x, n_ticket_y,n_ticket_colors = getCoordinatesTickets(normal_ticketData,ticket_names,listOfColors)
n_ticket_z = getTicketZValues(normal_ticketData)

```
